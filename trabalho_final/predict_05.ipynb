```python
from autoML import AutoML
from multiprocessing import Pool

pool = Pool(processes=10)
automl = AutoML(dsViaturas, 'y', min_x_y_correlation_rate=0.05, pool=pool, ds_name='viaturas_hiperparam', ngen=100)
automl.getResults()
```
